```python
crime_ctegori
```

13. Write a Python program to create a stacked bar plot with error bars.
Note: Use bottom to stack the women?s bars on top of the men?s bars.
Sample Data:

Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)
Men Standard deviation = (4, 3, 4, 1, 5)
Women Standard deviation = (3, 5, 2, 3, 3)
The code snippet gives the output shown in the following screenshot:

```python
data_bar = {'Group':['Group1','Group2','Group3','Group4','Group5'],'means_men':[22,30,35,35,26],'means_women':[25,32,30,35,29],'men_sd':[4,3,4,1,5],'women_sd':[3,5,2,3,3]}
means_sd = pd.DataFrame(data_bar)
means_sd_= means_sd[['Group','means_men','means_women']].copy()
l=['Group','Men','Women']
means_sd_.rename(columns=dict(zip(means_sd_.columns,l)),inplace=True)
```

```python
plt.figure(figsize=(10,5))
means_sd_.plot.bar(stacked=True,x='Group',color=['red','green'],yerr=[means_sd['men_sd'],means_sd['women_sd']])
plt.legend()
plt.show()
```
