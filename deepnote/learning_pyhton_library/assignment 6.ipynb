Q3. Draw the bar plot of the total count of the crimes occurring in a state year wise.

data preprocessing

```
state_crime['year'] = state_crime['year'].astype(dtype=str)
```

```
state_crime.head(2)
```

```
state_total_sum = state_crime.groupby(['state'])[['data_totals_violent_all']].sum().reset_index()
state_total_sum['totals_violent_mod'] = [(i%37000) for i in state_total_sum['data_totals_violent_all']]
state_total_sum
```

```
plt.figure(figsize=(30,10))
ax = sns.barplot(data=state_total_sum,x='state',y='totals_violent_mod',color='#eb52d4',gap=0.2,width=0.9)
ax.bar_label(ax.containers[0],fontsize=10)
plt.xticks(rotation=90)
plt.title(loc='left',label='all-year-total crime vs state',fontdict={'color':'#5c07f0','fontsize':15,'fontweight':'bold'})
plt.text(25,38000,'y-axis multiply with 37000 to get actual data of y-axis',bbox={'facecolor':'red','alpha':0.2})
plt.grid(alpha=0.4,color='r')
#plt.savefig('bar',dpi=500)
plt.show()
```

Q4. Choose the best plot to visualize the crimes distribution with their categories in each state and
answer that which state tops in which crime based upon the plot.

```
state_crime.head(2)
```

```
crime_ctegori = state_crime.groupby(['state'])[['data_totals_violent_assault','data_totals_violent_murder','data_totals_violent_rape','data_totals_violent_robbery']].sum().reset_index()
```

```
crime_ctegori['data_totals_violent_assault'] = [i%24000 for i in crime_ctegori['data_totals_violent_assault']]
crime_ctegori['data_totals_violent_murder'] = [i%500 for i in crime_ctegori['data_totals_violent_murder']]
crime_ctegori['data_totals_violent_rape'] = [i%7100 for i in crime_ctegori['data_totals_violent_rape']]
crime_ctegori['data_totals_violent_robbery'] = [i%4400 for i in crime_ctegori['data_totals_violent_robbery']]
```

```
crime_ctegori
```

```
a = crime_ctegori.loc[crime_ctegori['data_totals_violent_assault']== 23647]
m = crime_ctegori.loc[crime_ctegori['data_totals_violent_murder']== 497]
r = crime_ctegori.loc[crime_ctegori['data_totals_violent_rape']== 7086]
rob = crime_ctegori.loc[crime_ctegori['data_totals_violent_robbery']== 4283]
print(a)
print(m)
print(r)
print(rob)
```

```
plt.figure(figsize=(20,8))
sns.set_style("whitegrid")
ass = sns.lineplot(data=crime_ctegori,x='state',y='data_totals_violent_assault',label='assult',)
mur = sns.lineplot(data=crime_ctegori,x='state',y='data_totals_violent_murder',label='murder')
rap = sns.lineplot(data=crime_ctegori,x='state',y='data_totals_violent_rape',label='rape')
rob = sns.lineplot(data=crime_ctegori,x='state',y='data_totals_violent_robbery',label='robbery')
ass.plot('United States',23647,"*",markersize=10,color='r',label='assult-max * 2400')
mur.plot('Kansas',497,"o",markersize=10,color='r',label='murder-max * 500')
rap.plot('Oklahoma',7086,"+",markersize=20,color='r',label='rape-max * 7100')
rob.plot('Missouri',4285,"+",markersize=30,color='#0c17f0',linewidth=5,label='robbery_max * 4400')
plt.legend(loc='upper left')
plt.xticks(rotation=90)
plt.ylim([0,24000])
plt.grid(alpha=0.4,color='pink')
plt.show()
```

Q7. Draw a heatmap diagram to visualize the relationships between various crimes with other factors.

```
state_crime_violence_type = state_crime[['data_totals_violent_assault','data_totals_violent_murder','data_totals_violent_rape','data_totals_violent_robbery']].copy()
```

```
state_crime_type = state_crime[['data_totals_property_burglary','data_totals_property_larceny','data_totals_property_motor']] 
```

```
state_crime.head(2)
```

```
heatmap_generate = pd.merge(state_crime_violence_type,state_crime_type,left_index=True,right_index=True)
columns = ['Assault','Murder','Rape','Robbery','Burglary','Larceny','Motor']
heatmap_generate.rename(columns=dict(zip(heatmap_generate.columns,columns)),inplace=True)
heatmap_generate = heatmap_generate.corr()
```

```
heatmap_generate.head(2)
```

```
plt.figure(figsize=(10,10))
sns.set_style()
sns.heatmap(heatmap_generate,cmap='Accent',linewidths=0.5,cbar=True,annot=True,fmt='.2f')
plt.title('heat commparision between all property of assult and property',fontdict={'fontsize':10,'color':'blue','fontweight':"bold"})
plt.show()
```
